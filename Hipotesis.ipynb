2. Edad Promedio
* H0: La edad promedio de los clientes que interactúan con el nuevo proceso es igual a la del proceso anterior.
* H1: La edad promedio de los clientes que interactúan con el nuevo proceso es diferente de la del proceso anterior.

```python
#Unimos los dos df con id de cliente en común
df_users_beh_ex = pd.merge(df_users, df_beh_ex, on='client_id', how='inner')
df_users_beh_ex
```

```python
df_users_beh_ex = df_users_beh_ex.drop(['visitor_id', 'visit_id', 'Variation_y'], axis=1)
```

```python
df_users_beh_ex_u = df_users_beh_ex.drop_duplicates(subset='client_id', keep='first')
```

```python
df_users_beh_ex
```

Diferencia de promedios de edad entre antes y despues

```python
# H0: df_control [clnt_age] == df_test [clnt_age]
# H1: df_control [clnt_age] != df_test [clnt_age]

df_age_c= df_users_beh_ex_u[df_users_beh_ex_u["Variation_x"]=="Control"] ["clnt_age"]
df_age_t= df_users_beh_ex_u[df_users_beh_ex_u["Variation_x"]=="Test"] ["clnt_age"]

stat, p_value = st.ttest_ind(df_age_c, df_age_t, equal_var=False)
print ("stat: ", stat)
print ("p_value: ", p_value)

if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypotesis")
```
